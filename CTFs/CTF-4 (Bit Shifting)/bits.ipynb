```python
with open("bits.txt", "rb") as f:
    # Read the content of the file
    text = f.read()
    
print(text)
binary_representation=""
for char in text:
    print(bin(char)[2:].zfill(8))
    binary_representation += bin(char)[2:].zfill(8)
print(binary_representation)
```

```python
x="10100"
x='0'+x[:-1]
print(x)
```

```python
shifted_representation = binary_representation
for i in range(1,20):
    shifted_representation = shifted_representation[1:]+"0"
    # Convert the shifted binary representation back to text
    text = ''.join(chr(int(shifted_representation[j:j+8], 2)) for j in range(0, len(shifted_representation), 8))
    
    print(f"left Shift {i}: {text}")
```
