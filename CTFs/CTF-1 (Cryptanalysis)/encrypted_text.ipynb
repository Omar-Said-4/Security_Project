```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
def encrypt_letter(letter):
    if letter.isalpha():
        if letter.islower():
            return chr(((ord(letter) - 97 - 25) % 26) + 97)  # Lowercase letters
        else:
            return chr(((ord(letter) - 65 - 25) % 26) + 65)  # Uppercase letters
    else:
        return letter
```

```python
with open('encrypted_text.txt', 'r') as file:
    text = file.read()
letter_counts = {}
total_letters = 0
for letter in text:
    if letter.isalpha():  # Only consider letters
        letter_counts[letter] = letter_counts.get(letter, 0) + 1
        total_letters += 1

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.bar(letter_counts.keys(), letter_counts.values(), color='skyblue')
plt.xlabel('Letters')
plt.ylabel('Frequency')
plt.title('Frequency of Each Letter in the Encrypted Text')
plt.show()
```

```python
letter_percentages = [(letter, (count / total_letters) * 100) for letter, count in letter_counts.items()]

# Sort the list of tuples by percentage in descending order
letter_percentages.sort(key=lambda x: x[1], reverse=True)
print(letter_percentages)
```
